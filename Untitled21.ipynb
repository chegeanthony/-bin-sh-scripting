```python
# Assuming you've identified which features are most relevant based on your EDA
# Here's a list of features you've decided to keep for modeling
selected_features = [
    'Inflight wifi service',
    'Ease of Online booking',
    'Seat comfort',
    # Add other features that you found relevant from your EDA
]

# Create a new DataFrame with only the selected features
X = data[selected_features]

# Assuming 'Satisfied' is the target variable you want to predict
y = data['Satisfied']  # Replace with your actual target variable name
```

```python
# Example using pandas get_dummies for one-hot encoding
X_encoded = pd.get_dummies(X, drop_first=True)

# If you need to use LabelEncoder for ordinal encoding or encoding the target variable:
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the features and transform
X_scaled = scaler.fit_transform(X_encoded)
```

```python
from sklearn.model_selection import train_test_split

# Split the data - 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Initialize the models with default parameters
decision_tree = DecisionTreeClassifier(random_state=42)
random_forest = RandomForestClassifier(random_state=42)
```

```python
print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
```

```python
# Re-splitting the data to ensure alignment
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("Shape of X_train after re-splitting:", X_train.shape)
print("Shape of y_train after re-splitting:", y_train.shape)
```

```python
print("Number of NaN values in y_train after re-splitting:", y_train.isna().sum())
```

```python
# Initialize the Logistic Regression model
log_reg = LogisticRegression()

# Fit the model on the training data
log_reg.fit(X_train, y_train)
```

```python
# Train the decision tree model
decision_tree.fit(X_train, y_train)
# Train the random forest model
random_forest.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Initialize the grid search model
grid_search = GridSearchCV(estimator=random_forest, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")

# Train the random forest model with the best parameters
random_forest_best = RandomForestClassifier(**best_params, random_state=42)
random_forest_best.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predictions on test set for decision tree
y_pred_decision_tree = decision_tree.predict(X_test)

# Calculate metrics for decision tree
accuracy_dt = accuracy_score(y_test, y_pred_decision_tree)
precision_dt = precision_score(y_test, y_pred_decision_tree, average='weighted')
recall_dt = recall_score(y_test, y_pred_decision_tree, average='weighted')
f1_dt = f1_score(y_test, y_pred_decision_tree, average='weighted')

print(f"Decision Tree - Accuracy: {accuracy_dt}, Precision: {precision_dt}, Recall: {recall_dt}, F1 Score: {f1_dt}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predictions on test set for random forest
y_pred_random_forest = random_forest_best.predict(X_test)

# Ensure that y_test and predictions are in the correct binary format
# If your labels are strings like 'N' and 'Y', then specify pos_label='Y' (or 'N' for the negative class)

# Calculate metrics for random forest
accuracy_rf = accuracy_score(y_test, y_pred_random_forest)
precision_rf = precision_score(y_test, y_pred_random_forest, pos_label='Y')  # Replace 'Y' with your positive class label
recall_rf = recall_score(y_test, y_pred_random_forest, pos_label='Y')        # Replace 'Y' with your positive class label
f1_rf = f1_score(y_test, y_pred_random_forest, pos_label='Y')                # Replace 'Y' with your positive class label

print(f"Random Forest - Accuracy: {accuracy_rf}, Precision: {precision_rf}, Recall: {recall_rf}, F1 Score: {f1_rf}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate confusion matrix for decision tree
conf_matrix_dt = confusion_matrix(y_test, y_pred_decision_tree)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_dt, annot=True, fmt='g', cmap='viridis')
plt.title('Confusion Matrix for Decision Tree')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from scipy.stats import ttest_rel

# Use the cross-validated scores from your decision tree and random forest models
dt_scores = [0.81733314, 0.81733314, 0.81656321, 0.81717998]  # Replace with your actual scores
rf_scores = [0.81729992, 0.81729992, 0.81734712, 0.81711613]  # Replace with your actual scores

# Perform the paired t-test
t_stat, p_value = ttest_rel(dt_scores, rf_scores)

print(f"t-statistic: {t_stat}, p-value: {p_value}")
if p_value < 0.05:
    print("There's a significant difference between the models.")
else:
    print("There's no significant difference between the models.")
```
